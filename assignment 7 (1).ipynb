```python
import pandas as pd
```

```python
dt = pd.read_csv('order_items_dataset.csv')
```

```python
dt.iloc[:30000,:]
```

```python
dt['Quantity'] = dt.groupby(['order_id', 'product_id', 'seller_id', 'shipping_limit_date', 'price', 'freight_value'])['product_id'].transform('count')
```

```python
dt = dt.drop(columns=['order_item_id'])
```

```python
dt.shape
```

```python
print("total no. of unique order : ", len(dt['order_id'].unique()))
print("total no. of unique products : ", len(dt['product_id'].unique()))
```

```python
dt.describe(include='all')
```

```python
pt = dt.drop_duplicates(subset=['order_id', 'product_id'])
```

```python
product_counts = dt['product_id'].value_counts()
print('product_id value counts : \n\nproduct_id\t\t\t\tcount\n\n',product_counts)
```

```python
min_threshold = 10
popular_products = product_counts[product_counts >= min_threshold]
print("Products purchased at least 10 times : \n\nproduct_id\t\t\t\tcount\n\n",popular_products)
```

```python
avg_prod = dt.groupby('order_id')['Quantity'].sum().mean()
print("the average product purchased : ",avg_prod)
```
